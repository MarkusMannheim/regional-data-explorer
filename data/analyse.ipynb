```python
import pandas as pd, numpy as np

area_data = pd.read_csv("./area_data.csv")
area_data.dropna(inplace=True)
area_data = area_data[area_data["link"].apply(lambda x: ".csv" in x)]
area_data["area"] = area_data["area"].apply(lambda x: x.replace(" (ACT)", ""))

for i in area_data.index:
    data = pd.read_csv(
        area_data.at[i, "link"]
    )
    for j in data.index:
        area_data.at[i, data.at[j, "Description"]] = data.loc[j].dropna()[-1]
    area_data.at[i, "past_population"] = data.at[0, "2015"]

area_data.to_csv("./sa2_data.csv", index=False)
```

```python
data = area_data.loc[:, "code":"area"]
data["population"] = area_data[area_data.columns[3]]
data["population density"] = area_data[area_data.columns[4]]
data["median age"] = area_data[area_data.columns[9]]
data["children under 15"] = area_data[area_data.columns[102:105]].sum(axis=1) / 100
data["people over 64"] = area_data[area_data.columns[115:120]].sum(axis=1) / 100
data["fertility rate"] = area_data[area_data.columns[121]] / 100
data["Indigenous"] = area_data[area_data.columns[129]] / 100
data["religious"] = 1 - (area_data[area_data.columns[147]] / 100)
data["born overseas"] = area_data[area_data.columns[139]] / 100
data["born in Oceania"] = area_data[area_data.columns[130]] / 100
data["born in north-west Europe"] = area_data[area_data.columns[131]] / 100
data["born in southern & eastern Europe"] = area_data[area_data.columns[132]] / 100
data["born in north Africa & Middle East"] = area_data[area_data.columns[133]] / 100
data["born in south-east Asia"] = area_data[area_data.columns[134]] / 100
data["born in north-east Asia"] = area_data[area_data.columns[135]] / 100
data["born in southern & central Asia"] = area_data[area_data.columns[136]] / 100
data["born in Americas"] = area_data[area_data.columns[137]] / 100
data["born in sub-Saharan Africa"] = area_data[area_data.columns[138]] / 100
data["median house price"] = area_data[area_data.columns[248]]
data["median income"] = area_data[area_data.columns[312]]
data["low income (bottom quartile)"] = area_data[area_data.columns[322]] / 100
data["high income (top quartile)"] = area_data[area_data.columns[325]] / 100
data["postgraduates"] = area_data[area_data.columns[391]] / 100
data["university graduates"] = area_data[area_data.columns[391:394]].sum(axis=1) / 100
data["jobs: managers"] = area_data[area_data.columns[421]] / 100
data["jobs: professionals"] = area_data[area_data.columns[422]] / 100
data["jobs: technicians & trades"] = area_data[area_data.columns[423]] / 100
data["jobs: community work"] = area_data[area_data.columns[424]] / 100
data["jobs: clerical"] = area_data[area_data.columns[425]] / 100
data["jobs: sales"] = area_data[area_data.columns[426]] / 100
data["jobs: machinery & drivers"] = area_data[area_data.columns[427]] / 100
data["jobs: labourers"] = area_data[area_data.columns[428]] / 100
data["home owners (outright)"] = area_data[area_data.columns[505]] / 100
data["home owners (mortgage)"] = area_data[area_data.columns[506]] / 100
data["renters"] = area_data[area_data.columns[507]] / 100
data["detached houses"] = area_data[area_data.columns[500]] / area_data[area_data.columns[504]]
data["artificial surfaces"] = area_data[area_data.columns[601]] / area_data[area_data.columns[583]]
data.to_csv("./data.csv", index=False)
geo = pd.read_csv("data_geocoded.csv")
data_clean = data.set_index("code").join(geo[["code", "lng", "lat"]].set_index("code"))
data_clean = data_clean[data_clean["population"] > 149]
data_clean.to_csv("./data_clean.csv")
```
