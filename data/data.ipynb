```python
# required modules
import pandas as pd, numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time

# def setup_data():
#     return pd.DataFrame(
#     )

def setup_scraper():
    chrome_options = Options()
    chrome_options.headless = True
    return webdriver.Chrome(options=chrome_options)

def get_areas():
    data = pd.read_excel(
        "./SA2s.xlsx",
        usecols=[0, 1, 10, 14],
        names=["code", "area", "state", "size"]
    )
    data = data[(data["state"] == "8") & (data["size"] > 0)]
    return data[["code", "area"]]

def setup_data():
    return pd.DataFrame(columns=["code", "area", "link"])    

# start program!
driver = setup_scraper()
areas = get_areas()
area_data = setup_data()

# gather data links
for i in areas.index:
    print("gathering data for", areas.at[i, "area"], "...")
    driver.get(f"https://dbr.abs.gov.au/region.html?lyr=sa2&rgn={areas.at[i, 'code']}")
    download_link = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "dloadcsv"))
    )
    time.sleep(2)
    data_link = download_link.get_attribute("href")
    area_data.loc[len(area_data)] = [
        areas.at[i, "code"],
        areas.at[i, "area"],
        data_link
    ]

driver.close()

area_data
```

```python
area_data.to_csv("./area_data.csv", index=False)
```
